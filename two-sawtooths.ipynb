```python
from scipy import signal
import matplotlib.pyplot as plt
import numpy as np
import sounddevice as sd
import time
```

```python
n_samples = 4410
hz = 50
t = np.linspace(0, 1, n_samples)
waveform = signal.sawtooth(2 * np.pi * hz * t)
plt.plot(t, waveform)
```

```python
n_samples2 = 4410
hz2 = 15
t2 = np.linspace(0, 1, n_samples2)
waveform2 = signal.sawtooth(2 * np.pi * hz2 * t2)
plt.plot(t2, waveform2)
```

```python
combined_waveform = waveform + waveform2
plt.plot(t2, combined_waveform)
```

```python
# Samples per second
sps = 44100 #samples per second, 44.1 khz, cd quality

# Play the waveform

combined_waveform = waveform + waveform2
sd.play(combined_waveform, sps, loop=True)
time.sleep(2)
sd.stop()
```
